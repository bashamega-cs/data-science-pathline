```python
import pandas as pd
dtype = {
    "id": "int64",

    "NAME": "string",

    "host id": "Int64",
    "host_identity_verified": "string",
    "host name": "string",

    "neighbourhood group": "string",
    "neighbourhood": "string",

    "lat": "Float64",
    "long": "Float64",

    "country": "string",
    "country code": "string",

    "instant_bookable": "boolean",
    "cancellation_policy": "string",
    "room type": "string",

    "Construction year": "Int64",

    "price": "string",
    "service fee": "string",

    "minimum nights": "Int64",

    "number of reviews": "Int64",
    "reviews per month": "Float64",
    "review rate number": "Int64",

    "calculated host listings count": "Int64",
    "availability 365": "Int64",

    "house_rules": "string",
    "license": "string",
}

df = pd.read_csv(
    "Airbnb_Open_Data.csv",
    dtype=dtype,
    parse_dates=["last review"]  # optional but recommended
)
```

```python
# Identify columns that are of object (string) type and fill them with 'Unknown'
columns_to_fill = df.select_dtypes(include=['object', 'string']).columns.tolist()

# Define a default fill value for all these columns
fill_value = "Unknown"

# Fill nulls only for these columns
df[columns_to_fill] = df[columns_to_fill].fillna(fill_value)
```

```python
# Identify columns that are of object (boolean) type and fill them with 'FALSE'
columns_to_fill = df.select_dtypes(include=['object', 'boolean']).columns.tolist()

# Define a default fill value for all these columns
fill_value = False

# Fill nulls only for these columns
df[columns_to_fill] = df[columns_to_fill].fillna(fill_value)
```

```python
df.to_csv("output.csv", index=False, encoding="utf-8")
```
