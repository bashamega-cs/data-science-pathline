```python
import pandas as pd
dtype = {
    "id": "int64",

    "NAME": "string",

    "host id": "Int64",
    "host_identity_verified": "string",
    "host name": "string",

    "neighbourhood group": "string",
    "neighbourhood": "string",

    "lat": "Float64",
    "long": "Float64",

    "country": "string",
    "country code": "string",

    "instant_bookable": "boolean",
    "cancellation_policy": "string",
    "room type": "string",

    "Construction year": "Int64",

    "price": "string",
    "service fee": "string",

    "minimum nights": "Int64",

    "number of reviews": "Int64",
    "reviews per month": "Float64",
    "review rate number": "Int64",

    "calculated host listings count": "Int64",
    "availability 365": "Int64",

    "house_rules": "string",
    "license": "string",
}

df = pd.read_csv(
    "Airbnb_Open_Data.csv",
    dtype=dtype,
    parse_dates=["last review"]  # optional but recommended
)
```

```python

# Fill nulls for all string and boolean columns
fill_dict = {}

# Strings → "Unknown"
for col in df.select_dtypes(include=['object', 'string']).columns:
    fill_dict[col] = "Unknown"

# Booleans → False
for col in df.select_dtypes(include=['boolean']).columns:
    fill_dict[col] = False

# Number -> 0
for col in df.select_dtypes(include=['Int64', 'Float64']).columns:
    fill_dict[col] = 0

# Apply all fills at once
df.fillna(value=fill_dict, inplace=True)
```

```python
df.to_csv("output.csv", index=False, encoding="utf-8")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Copy your column
ratings = df['number of reviews']

# Set all ratings larger than 100 to 100
ratings = ratings.apply(lambda x: 100 if x > 100 else x)

# Bin ratings into ranges of 10
bins = list(range(0, 101, 10))  # 0-9, 10-19, ..., 90-100
labels = [f"{b}-{b+9}" for b in bins[:-1]]
ratings_binned = pd.cut(ratings, bins=bins, labels=labels, right=False)

# Count how many ratings in each bin
bin_counts = ratings_binned.value_counts().sort_index()

# Plot the bar chart
plt.figure(figsize=(12,6))
sns.barplot(x=bin_counts.index, y=bin_counts.values)
plt.xticks(rotation=45)  # Rotate x-axis labels so they don’t overlap
plt.xlabel("Rating Range")
plt.ylabel("Number of Ratings")
plt.title("Number of Ratings per Rating Range (bins of 10, max 100)")
plt.show()
```
